Дослідіть токенізацію за допомогою tiktoken - швидкого токенізатора з відкритим кодом від OpenAI
Дивіться [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) для додаткових прикладів.


````python
# ВПРАВА:
# 1. Спочатку запустіть вправу як є
# 2. Змініть текст на будь-який промпт, який ви хочете використати, і перезапустіть, щоб побачити токени

import tiktoken

# Визначте промпт, який ви хочете токенізувати
text = f"""
Юпітер - п'ята планета від Сонця і найбільша \
в Сонячній системі. Це газовий гігант з масою \
в одну тисячну від маси Сонця, але в два з половиною \
рази більшою за масу всіх інших планет Сонячної системи разом узятих. \
Юпітер - один з найяскравіших об'єктів, видимих неозброєним оком \
на нічному небі, і був відомий стародавнім цивілізаціям ще \
до писаної історії. Він названий на честь римського бога Юпітера.[19] \
При спостереженні з Землі Юпітер може бути настільки яскравим, що його відбите \
світло може відкидати видимі тіні,[20] і в середньому він є третім за яскравістю \
природним об'єктом на нічному небі після Місяця та Венери.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Вправа 2: Перевірка налаштування ключа Github Models

Запустіть код нижче, щоб перевірити, чи правильно налаштована кінцева точка Github Models. Код просто випробовує простий базовий промпт і перевіряє завершення. Вхідний текст `о скажи, чи бачиш ти` має завершитись відповідно до контексту.


````python
!pip install azure-ai-inference
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

def get_completion(prompt, client, model_name, temperature=1.0, max_tokens=1000, top_p=1.0):
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "Ви - корисний асистент.",
            },
            {
                "role": "user",
                "content": prompt,
            },
        ],
        model=model_name,
        temperature=temperature,
        max_tokens=max_tokens,
        top_p=top_p
    )
    return response.choices[0].message.content

## ---------- Виклик допоміжного методу

### 1. Встановіть основний вміст або текст промпту
text = f"""
о скажи, чи бачиш ти
"""

### 2. Використайте його в шаблоні промпту нижче
prompt = f"""
```{text}```
"""

## 3. Запустіть промпт
response = get_completion(prompt, client, model_name)
print(response)
````

### Вправа: Дослідіть свою інтуїцію
Наведені вище приклади дають вам шаблони, які ви можете використовувати для створення нових промптів (простих, складних, інструкційних тощо) - спробуйте створити інші вправи для дослідження деяких інших ідей, про які ми говорили, таких як приклади, підказки та інше.

````python
# Варіант 9: Створення промпту для аналізу дампів пам'яті x86/ARM.
# Виконав: Яценко Матвій

import tiktoken

class MemoryForensicsAnalyzer:
    def __init__(self, model_name="gpt-4o"):
        self.model = model_name
        self.encoder = tiktoken.encoding_for_model(model_name)

    def build_prompts(self):
        # 1. Простий промпт (Base Prompt)
        base = (
            "Проаналізуй цей дамп пам'яті архітектури x86/ARM. "
            "Знайди ознаки переповнення буфера та підозрілі стрибки виконання."
        )

        # 2. Складний системний промпт (Detailed System Prompt)
        system = (
            "Ти — провідний експерт із низькорівневої безпеки та форензики. "
            "Твоє завдання — автоматизований розбір дампів пам'яті (core dumps). "
            "Аналізуй стан регістрів, вказівники на стек та heap. "
            "Якщо виявлено аномалію, вказуй ймовірний вектор атаки (ROP, Shellcode injection). "
            "Відповідь надавай у форматі суворого технічного звіту з секціями: 'Об'єкт', 'Аномалія', 'Рівень ризику'."
        )
        return base.strip(), system.strip()

    def get_token_count(self, text):
        return len(self.encoder.encode(text))

    def calculate_efficiency(self, prompt):
        # Логіка оцінки якості (аналогічна оригіналу, але переписана)
        keywords = ["аналіз", "архітектура", "стек", "регістри", "звіт", "безпека", "аномалія", "формат"]
        score = sum(1 for word in keywords if word in prompt.lower())
        
        # Додаткові бали за структуру
        if len(prompt) > 150: score += 1
        if ":" in prompt: score += 1
        
        return min(score, 10)

def main():
    analyzer = MemoryForensicsAnalyzer()
    p_base, p_sys = analyzer.build_prompts()

    # Вивід результатів аналізу токенів
    print(f"--- Аналіз промптів для моделі {analyzer.model} ---")
    print(f"Base Prompt: {analyzer.get_token_count(p_base)} токенів")
    print(f"System Prompt: {analyzer.get_token_count(p_sys)} токенів")
    
    # Оцінка
    sys_score = analyzer.calculate_efficiency(p_sys)
    print(f"\nРейтинг системного промпту: {sys_score}/10")

    # Рекомендації щодо покращення (персоналізовані)
    suggestions = [
        "Додати підтримку специфічних форматів (ELF для Linux або PE для Windows).",
        "Впровадити 'Few-shot' приклади з реальними адресами пам'яті для кращого контексту.",
        "Чітко визначити обмеження: вказати моделі ігнорувати нульові сторінки пам'яті.",
        "Додати вимогу щодо виводу результатів у форматі JSON для подальшої автоматизації."
    ]

    print("\nШляхи вдосконалення інженерії промптів:")
    for i, step in enumerate(suggestions, 1):
        print(f"{i}. {step}")

if __name__ == "__main__":
    main()
````
